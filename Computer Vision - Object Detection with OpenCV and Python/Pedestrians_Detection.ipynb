### Import Libraries

```python
import cv2
import numpy as np
```

### Create Classifier

```python
body_classifier=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_fullbody.xml")
```

### Capture Video

```python
cap = cv2.VideoCapture('Video/People_Walking.mp4')
```

```python
# While Loop
while cap.isOpened():
    
    # Read the capture
    ret, frame = cap.read()
    
    # Pass the Frame to the Classifier
    bodies = body_classifier.detectMultiScale(frame, 1.2, 3)
    
    # if Statement
    if ret == True:
    
        # Bound Boxes to Identified Bodies
        for (x,y,w,h) in bodies:
            cv2.rectangle(frame,
                         (x,y),
                         (x+w, y+h),
                         (255,125,225),
                         5)
            cv2.imshow('Pedestrians', frame)
        
        # Exit with Esc button
        if cv2.waitKey(1) == 27:
            break
        
    # else Statement
    else:
        break
    
# Release the Capture & Destroy All Windows
cap.release()
cv2.destroyAllWindows()
```
